```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Convertir Dataframe a CSV
data.to_csv("Inmuebles2.csv")
data
```

```python
#Eliminemos la primera fila que contiene información sobre la base de datos 
data1= pd.read_csv('Inmuebles2.csv', skiprows=1, low_memory=False)
data1
```

```python
#Eliminar columnas innecesarias
data2= data1.drop(['0','Referencia'], axis=1)
data2
```

```python
#Verifiquemos la existencia de valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
#Se identifican valores nulos en Fecha de Alta, Tipo, Operación, Provincia, Superficie, Precio Venta, Fecha Venta y Vendedor 
```

```python
#Comencemos con la columna fecha alta 
#Sustituir valores nulos por un string en concreto 
data2['Fecha Alta'] = data2['Fecha Alta'].fillna("2004-01-01 00:00:00")
#Revisamos que nuestro método haya funcionado 
valores_nulos=data2.isnull().sum()
valores_nulos
#Concatenación de columnas 
fecha = data2["Fecha Alta"].fillna("2004-01-01 00:00:00")
```

```python
#Sigamos con la columna Tipo 
#Eliminamos los registros de las filas que contienen valores nulos para la columna Tipo
data2["Tipo"] = data2['Tipo'].fillna("Unknown")
#Revisamos que nuestro método haya funcionado 
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Tratemos el dato faltante de la columna Operación 
data2["Operación"] = data2['Operación'].fillna("Unknown")
#Revisamos que nuestro método haya funcionado 
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Continuemos con el tratamiento de datos nulos de la columna Provincia 
data2["Provincia"] = data2['Provincia'].fillna("Unknown")
#Revisamos que nuestro método haya funcionado 
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Continuemos con el tratamiento de datos nulos de la columna Superficie
#Sustituimos las superficies desconocidas con el identificador clave 1
data2['Superficie'] = data2['Superficie'].fillna(1)
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Continuemos con el tratamiento de datos nulos de la columna Precio Venta
data2['Precio Venta'] = data2['Precio Venta'].fillna(data2["Precio Venta"].mean())
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Eliminemos las columnas restantes Fecha Venta y Vendedor debido a su gran cantidad de NAs
data2= data2.dropna(0)
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Convertimos nuestro dataframe después de aplicado nuestro tratamiento de NA´s a un csv 
data2.to_csv("Inmuebles_2_clean.csv")
```
